```python
# Diagnostic + safe correlation heatmap (copy-paste into a new notebook cell)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

RESULTS_DIR = Path("../results/plots")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

# 1) show column dtypes so we can see which are numeric or not
print("Column dtypes:\n", df.dtypes)
print("\nNumeric columns ->", list(df.select_dtypes(include=[np.number]).columns))

# 2) pick numeric columns only
num_df = df.select_dtypes(include=[np.number]).copy()

if num_df.shape[1] < 2:
    print("\nNot enough numeric columns to compute a correlation matrix.")
else:
    # 3) compute correlation and plot heatmap
    corr = num_df.corr()
    print("\nCorrelation matrix:\n", corr)

    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
    plt.title("Correlation Heatmap (numeric cols only)")

    out_file = RESULTS_DIR / "correlation_heatmap.png"
    plt.savefig(out_file, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"\nSaved heatmap to: {out_file.resolve()}")
```

```python
# Step 5: Scatterplot Study Time vs Final Grade (with regression line)
plt.figure(figsize=(6,4))
sns.regplot(x="studytime", y="G3", data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title("Study Time vs Final Grades (G3) - Scatterplot with Regression Line")
plt.xlabel("Study Time (1=low, 4=high)")
plt.ylabel("Final Grade")

# Save plot
plt.savefig("../results/plots/studytime_vs_finalgrade_regression.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Step 6: Summary statistics

# Average grade by gender
avg_gender = df.groupby("sex")["G3"].mean()
print("Average Final Grade by Gender:\n", avg_gender, "\n")

# Average grade by studytime
avg_study = df.groupby("studytime")["G3"].mean()
print("Average Final Grade by Study Time:\n", avg_study, "\n")

# Categorize absences into low/high
df["absence_category"] = pd.cut(df["absences"], bins=[-1,3,10], labels=["Low Absences","High Absences"])
avg_absence = df.groupby("absence_category")["G3"].mean()
print("Average Final Grade by Absence Category:\n", avg_absence)
```

```python
# Step 7: Visualize Group Averages
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Average grade by gender
plt.figure(figsize=(5,4))
sns.barplot(x="sex", y="G3", data=df, estimator="mean", palette="Set2")
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.savefig("../results/plots/avg_grade_by_gender.png", dpi=300, bbox_inches="tight")
plt.show()

# 2. Average grade by studytime
plt.figure(figsize=(6,4))
sns.barplot(x="studytime", y="G3", data=df, estimator="mean", palette="viridis")
plt.title("Average Final Grade by Study Time")
plt.xlabel("Study Time (1=low, 4=high)")
plt.ylabel("Average Final Grade")
plt.savefig("../results/plots/avg_grade_by_studytime.png", dpi=300, bbox_inches="tight")
plt.show()

# 3. Average grade by absence category
plt.figure(figsize=(6,4))
sns.barplot(x="absence_category", y="G3", data=df, estimator="mean", palette="coolwarm")
plt.title("Average Final Grade by Absence Category")
plt.xlabel("Absence Category")
plt.ylabel("Average Final Grade")
plt.savefig("../results/plots/avg_grade_by_absence.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Student Performance Analysis - Summary

## Key Findings:
1. **Grade Distribution**: Most students scored between 12 and 15 in final grades.
2. **Gender vs Final Grades**: Both male and female students have similar performance, with no major differences.
3. **Study Time vs Final Grades**: Students who dedicated more time to studying showed slightly better grades.
4. **Absences vs Final Grades**: Higher absences generally led to lower grades.
5. **Correlation Heatmap**: Strong positive correlation between G1, G2, and G3 (previous and final grades).

## Conclusion:
- Study time and regular attendance improve final grades.
- Gender does not significantly impact performance.
- Early grades (G1, G2) are good predictors of final performance (G3).
